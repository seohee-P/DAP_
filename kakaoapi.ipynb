```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
subway = pd.read_csv('/content/drive/MyDrive/project/data/서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv', encoding = 'cp949')
```

```python
subway = subway['지하철역']
subway = set(subway)
sub={}
```

```python

X = []
Y = []
station = []
road_address = []
for subway_station in subway:
  url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
  params = {'query': (subway_station + "지하철"), 'category_group_code': "SW8"}
  headers = {"Authorization": "KakaoAK 6302172e3988baf79d41c56bf2c814e5"}
  places = requests.get(url, headers = headers, params=params).json()['documents']

  for place in places:
    if subway_station not in station:
      X.append(float(place['x']))
      Y.append(float(place['y']))
      station.append(subway_station)
      road_address.append(place['road_address_name'])
      
df = pd.DataFrame()
df['지하철역'] = station
df['X'] = X
df['Y'] = Y
df['도로명주소'] = road_address
df.to_csv('/content/drive/MyDrive/project/place.csv', encoding="utf-8", index=False)
```
